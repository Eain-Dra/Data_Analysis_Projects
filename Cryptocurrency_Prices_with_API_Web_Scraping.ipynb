```python
#Import all required modules
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
import os
import pandas as pd
from time import time
from time import sleep
import seaborn as sns

#Create a function to get data from website and make a dataframe and update dataframe everytime running function
def api_runner():
    global df

    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
    'start':'1',
    'limit':'10',
    'convert':'USD'
    }
    headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': '3be6c215-d99e-49a5-8077-5c6069aecfff',
    }

    session = Session()
    session.headers.update(headers)

    try:
        response = session.get(url, params=parameters)
        data = json.loads(response.text)
        print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
        print(e)

    df2=pd.json_normalize(data['data'])
    df2['timestamp']=pd.to_datetime('now')
    df3=pd.DataFrame(df2)
    df=pd.concat([df2,df3])
    df

#A loop for running function and set timer for sleep
for i in range(150):
    api_runner()
    print("API Completed successfully")
    sleep(60)
```

```python
#Group dataframe by required attributes
df4=df.groupby('name',sort=False)[['quote.USD.percent_change_1h','quote.USD.percent_change_24h','quote.USD.percent_change_7d','quote.USD.percent_change_30d','quote.USD.percent_change_60d','quote.USD.percent_change_90d']].sum()
df4
```

```python
#Change columns to rows
df5=df4.stack()
df5
```

```python
#Change to dataframe and add index
df6=df5.to_frame(name='values')
index=pd.Index(range(df5.count()))
df7=df6.set_index(index)
df7
```

```python
df7=df6.reset_index()
#Change column name and replace row name
df7=df7.rename(columns={'level_1':'%_change'})
df7['%_change']=df7['%_change'].replace(to_replace='quote.USD.percent_change_1h',value='1h')
df7['%_change']=df7['%_change'].replace(to_replace='quote.USD.percent_change_24h',value='24h')
df7['%_change']=df7['%_change'].replace(to_replace='quote.USD.percent_change_7d',value='7d')
df7['%_change']=df7['%_change'].replace(to_replace='quote.USD.percent_change_30d',value='30d')
df7['%_change']=df7['%_change'].replace(to_replace='quote.USD.percent_change_60d',value='60d')
df7['%_change']=df7['%_change'].replace(to_replace='quote.USD.percent_change_90d',value='90d')
df7
```

```python
sns.set_theme(style='darkgrid')
sns.catplot(x='%_change',y='values',hue='name',data=df7,kind='point')
```
